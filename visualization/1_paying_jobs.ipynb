```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file into a DataFrame
df = pd.read_csv(r'D:\DataNerd\SQL_Project_Data_Job_Analysis\project_sql\top_paying_jobs.csv')

# Display the first few rows to ensure it loaded correctly
print(df.head())
```

```python
# Create a color map
norm = plt.Normalize(df['salary_year_avg'].min(), df['salary_year_avg'].max())
cmap = plt.get_cmap('Blues')

# Create a color array based on the colormap for the salary_year_avg values
colors = [cmap(norm(value)) for value in df['salary_year_avg']]

# Plot
plt.figure(figsize=(10, 8))
bars = plt.barh(df['job_title'], df['salary_year_avg'], color=colors, edgecolor='black')
plt.xlabel('Average Salary')
plt.title('Top Paying Data Analyst Jobs in 2023')
plt.gca().invert_yaxis()  # Highest salary on top

# Show plot
plt.show()
```
