```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file into a DataFrame
df = pd.read_csv(r'D:\DataNerd\SQL_Project_Data_Job_Analysis\project_sql\top_paying_job_skills.csv')

# Display the first few rows to ensure it loaded correctly
print(df.head())
```

```python
# Count occurrences of each skill
skill_counts = df['skills'].value_counts()

# Convert to DataFrame
skill_counts_df = skill_counts.reset_index()
skill_counts_df.columns = ['skill', 'count']

# Display the DataFrame to check the counts
print(skill_counts_df.head())
```

```python
# Create a color map
norm = plt.Normalize(skill_counts_df['count'].min(), skill_counts_df['count'].max())
cmap = plt.get_cmap('Blues')  # You can choose other colormaps like 'coolwarm', 'viridis', etc.


# Create a color array based on the colormap
colors = [cmap(norm(value)) for value in skill_counts_df['count'][:10]]

# Plotting
plt.figure(figsize=(10, 6))
bars = plt.barh(skill_counts_df['skill'][:10], skill_counts_df['count'][:10], color=colors, edgecolor='black')
plt.xlabel('Frequency')
plt.title('Skill Count for Top 10 Paying Data Analyst Jobs in 2023')
plt.gca().invert_yaxis()  # Reverse the y-axis to have the highest count at the top
plt.tight_layout()

plt.show()
```
